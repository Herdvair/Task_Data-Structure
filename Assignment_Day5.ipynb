### Looping

```python
#looping for untuk mengidentifikasi tarif min dan max di tiap titik pelabuhan

df = pd.read_excel('titanic.xlsx')
nilai_plb = df['embarked'].dropna().unique() #menghilangkan nilai NaN di embarked

for titik_plb in nilai_plb:
    subset = df[df['embarked'] == titik_plb]
    max_fare = subset['fare'].max()
    min_fare = subset['fare'].min()
    
    print(f"Pelabuhan {titik_plb}:")
    print(f"fare tertinggi {max_fare}:")
    print(f"fare terendah {min_fare}:")
```

```python
df = pd.read_excel('titanic.xlsx')
data_penumpang = df[['name', 'age']]

def info_penumpang(nama, umur):
    print(f"Penumpang yang bernama {nama} berusia lebih dari {umur} tahun")

i = 0
while i < len(data_penumpang):
    nama_penumpang = data_penumpang.loc[i,'name']  #loc digunakan memilih baris dan kolom suatu data frame
    umur_penumpang = data_penumpang.loc[i,'age']
    
    if umur_penumpang > 36:
        info_penumpang(nama_penumpang, umur_penumpang)
    i+=1
```

### Lambda

```python
#numerikal
max_fare = df['fare'].max()
df[df['fare'].apply(lambda x: x == max_fare)]
```

```python
#kategorikal
df[df['sex'].apply(lambda x: x == 'female')]
```

- Pada lambda function di data numerikal mengidentifikasi penumpang yang membayar tarif termahal/tertinggi hanya sebanyak 4 orang
- Pada lambda function di data kategorikal mengidentifikasi jenis kelamin wanita dari dataframe "sex", hasilnya penumpang kapal titanic yang berjenis kelamin wanita itu sebanyak 475 orang yang artinya penumpang wanita lebih sedikit dibandingkan penumpang pria
